# oscilador armonico

```python
%matplotlib inline #se importan librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
#grafica de teta contra tiempo x como tiempo y y cmo teta
#grafica de x como tiempo y x como v
```

```python
h=.0001 #sera el incremento que tendra el tiempo
w=5 #esta dada 
tmax=10 #el tiempo maximo a calcular
t=np.arange(0,tmax,h) #para darle valores a el arreglo de tiempo
N=len(t) #cuantos tiempos habra
```

```python
v=np.zeros(N) #inicializo los valores de velocidad con la misma cantidad de elementos que tiempo
v[0]=.5 #doy valor al primer elemento de velocidad
```

```python
teta=np.zeros(N) #inicializo los valores de teta con la misma cantidad de elementos que tiempo

```

```python
for i in range (N-1): #ciclo for para realizar las operaciones para obtener teta y velocidad
   
    v[i+1]=v[i]+h*(-w**2*teta[i])
    teta[i+1]=teta[i]+h*v[i]    
    
   
```

# grafica de tiempo contra teta

```python
plt.plot(t,teta)
```


## grafica de velocidad contra teta

```python
plt.plot(v,teta)
```
