# uso y definicion de funciones

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=np.linspace(-10,10,100)
x2=np.linspace(-5,5,100)
y=np.cos(x)*5
y2=np.cos(x2)
plt.plot(x,y)
plt.plot(x2,y2)
```

```python
#opcion 1 para definir funciones
fx=lambda x: x**2*np.cos(x)-3*x**3
```

```python
fx(x)
```

```python
#opcion 2 para definir funciones
def fx2(x,y):
    z=x*y*np.cos(x/y)-3*x**3
    return z
```

```python
plt.plot(x,fx2(x,y))
```

```python
def recta(x,m,b):
    return m*x+b
```

```python
x=np.linspace(-5,5,20)
x2=np.linspace(-8,3,10)
x3=np.linspace(-4,4,30)
plt.plot(x,recta(x,3,5))
plt.plot(x2,recta(x2,2,9))
plt.plot(x3,recta(x3,3,4))
```

```python
def recta2(x,m,b=0):
    return m*x+b
```

```python
x=np.linspace(-5,5,20)
x2=np.linspace(-5,5,20)
x3=np.linspace(-5,5,20)
x4=np.linspace(-5,5,20)
x5=np.linspace(-5,5,20)
plt.plot(x,recta2(x,1))
plt.plot(x2,recta2(x2,2))
plt.plot(x3,recta2(x3,3))
plt.plot(x4,recta2(x4,4))
plt.plot(x5,recta2(x5,5))
```

```python
def recta3(x,m,b=0):
    return m*x+b
```

```python
for m in range (-5,5):
    x=np.linspace(-50,50,100)
    plt.plot(x,recta3(x,m))
```
