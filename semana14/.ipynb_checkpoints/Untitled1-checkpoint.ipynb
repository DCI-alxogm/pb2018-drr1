# oscilador armonico

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#grafica de teta contra tiempo x como tiempo y y cmo teta
#grafica de x como tiempo y x como v
```

```python
h=.001
w=5
t=np.arange(0,1,h)
N=len(t)
```

```python
v=np.zeros(N)
v[0]=.5
```

```python
teta=np.zeros(N)

```

```python
for i in range (N-1):
   
    v[i+1]=v[i]+h*(-w**2*teta[i])
    teta[i+1]=teta[i]+h*v[i]    
    
   
```

```python
plt.plot(t,teta)
```

```python
plt.plot(v,teta)
```
